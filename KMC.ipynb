##KMC task - 1+1D SOS surface

We model a system consisting of a 1D surface of length $L$ where atoms can attach to the surface at a constant rate $F$, and detach only from the peaks (no overhangs). This representation of a surface is referred to as a solid-on-solid (SOS) surface.

We will use periodic boundary conditions. The particles interact via bonds with neighboring particles.
An important measure of the energy associated with a particle is thus its number of neighbors:

```python
def periodic_transform(site, L):
    """
    Transformes a coordinate according to periodic boundary conditions
    for a system of size L.
    """
    
    return site % L

def number_of_neighbors(site, surface_heights, L):
    
    prev_site = periodic_transform(site - 1, L)
    next_site = periodic_transform(site + 1, L)
    
    prev_height = surface_heights[prev_site]
    next_height = surface_heights[next_site]
    
    height = surface_heights[site]
    
    neighbors = 1 #we are allways connected below
    
    if prev_height >= height:
        neighbors += 1
    if next_height >= height:
        neighbors += 1
    
    return neighbors        
```

The rate is given by $r_d = r_0 \exp(-n J/k_bT)$, where $n$ is the number of neihbors, $r_0$ is a rate constant, $J$ is the particle bond energy, $k_b$ is the Boltzmann constant, and $T$ is the temperature. In reduces units $r_d^\ast = r_d/r_0$ and $T^\ast = k_bT/J$ becomes

$$r_d^\ast = \exp(-n/T^\ast).$$ 

For stability, we introduce an energy shift of $2/T$, which normalizes the values of the rates without changing the dynamics:

$$r_d^\ast \to \exp(-(n-2)/T^\ast).$$ 

```python
def calculate_detachment_rate_from_neighbors(n, T):
    """
    Returns the detachment rate for n neighbors in units of r0 exp[2/T]
    at a given reduced temperature T.
    """
    
    return exp(-(n-2.)/T)

def calculate_detachment_rate(site, surface_heights, L, all_detachment_rates):
    """
    Returns the detachment rate at surface site [site] in units of r0
    using efficient calls to a precalculated vector.
    This way we save alot of exponential function calls.
    """
   
    n = number_of_neighbors(site, surface_heights, L)
    return all_detachment_rates[n-1]

def setup_all_detachment_rates_list(T):
    """
    returns a list of all detatchment rates which
    can be indices by the number of neighbors - 1.
    """
    
    return [calculate_detachment_rate_from_neighbors(1, T), 
            calculate_detachment_rate_from_neighbors(2, T),
            calculate_detachment_rate_from_neighbors(3, T)]
```

The probability of an event $r_e$ occuring is proportional to the rate of the event, that is

$$P(e) = r_e/r_t,$$

where 

$$r_t = \sum_e r_e,$$

is the sum of the rates of all events.

The time step in KMC is described as the "mean escape time" from the current state, and is equal to 

$$\delta t = -\log(\chi)/r_t,$$

where $\chi\in(0,1)$ is a uniform random number.

```python
def calculate_time_step(rate_sum):
    """
    Returns the time step for a given rate sum.
    """
    
    return -log(random.random())/rate_sum
    
```

We can now construct a Monte-Carlo algorithm which select an event at random with a probability weighted by its rate.
This is referred to as Kinetic Monte-Carlo (KMC). 

The first step is to construct a list of all events that can happen and their rates (in this case 2L events). 
We use that the first $L$ events represent deposition events at all surface sites, 
and the last $L$ elements represents the corresponding detachment events.

```python
def make_event_rate_list(L, surface_heights, F, all_detachment_rates):
    """
    Returns a list of the rates of all events that can occur.
    """
    
    rates = zeros(2*L)
    
    for site in range(L):                   #loop over all sites
        detachment_rate = calculate_detachment_rate(site, 
                                                    surface_heights,  
                                                    L,
                                                    all_detachment_rates)
        
        rates[site] = F                     #the first half is attachment rates
        rates[site + L] = detachment_rate   #the second half is detachment rates
    
    return rates
```

In order to select a random event with a probability proportional to its rate, we must first construct the cumulative sum list $\mathbf{r}_c$ such that

$$r_c(i) = \sum_{j=0}^i r_i$$.

The last element is thus equal to the sum over all rates $r_t$. 

```python
def make_cumulative_rate_list(rate_list, L):
    """
    Returns the cumulative rate list such that element i is equal to 
    the sum of all elements in rate_list up to and including i.
    """
    
    return cumsum(rate_list)
```

We can then select a random number $R\in[0, r_t)$, and locate the element in the cumulative sum which satisfies

$$r_c(i-1) \le R < r_c(i).$$


```python
def locate_in_cumulative_rates(R, cumulative_rate_list):
    """
    Returns the index i which satisfies 
    cumulative_rate_list[i-1] <= R < cumulative_rate_list[i].
    """
    
    for index, value in enumerate(cumulative_rate_list):
        if R < cumulative_rate_list[index]:
            return index
    
```

The index $i$ of this element will correspond to the index of the event that has been selected to occur.

```python
def select_event(L, surface_heights, F, all_detachment_rates):
    """
    Returns the index of a randomly selected event with a probability proportional to its rate. 
    And also the associated time step.
    """
    
    rate_list = make_event_rate_list(L, surface_heights, F, all_detachment_rates)
    cumulative_rate_list = make_cumulative_rate_list(rate_list, L)
    
    rate_sum = cumulative_rate_list[-1] #the last element of the cumulative list is the sum of all elements
    
    time_step = calculate_time_step(rate_sum)
    
    R = random.random()*rate_sum
    
    selected_event = locate_in_cumulative_rates(R, cumulative_rate_list)
    
    return selected_event, time_step
    
```

We now need only to translate a selected event into a consequence for the surface. 

```python
def execute_event(selected_event, L, surface_heights):
    """
    Executes the consequence of a selected event,
    either increasing or decreasing the surface height of the 
    respective event's site by one unit.
    """
    
    #This means that we should deposit a particle at site [selected_event]
    if selected_event < L:
        height_change = 1
        site = selected_event
        
    #Else we are detaching a particle from the site [selected_event] - L
    else:
        height_change = -1
        site = selected_event - L
       
    surface_heights[site] += height_change
    
```

Putting it all together yields the following complete simulation implementation

```python
def simulate_surface(surface_heights, L, N, T, F):
    """
    Simulates an initial surface [surface_heights] of length [L] for [N] steps
    at a reduced temperature [T] with a deposition rate [F].
    """
    
    all_surfaces = zeros([N, L])
    all_times = zeros(N)
    total_time = 0
    
    all_detachment_rates = setup_all_detachment_rates_list(T)
    
    for cycle in range(N):
        
        selected_event, time_step = select_event(L, surface_heights, F, all_detachment_rates)
        execute_event(selected_event, L, surface_heights)
            
        all_surfaces[cycle, :] = surface_heights
        
        total_time += time_step
        all_times[cycle] = total_time
    
    return all_surfaces, all_times
```

Now, all we need to do is to set a number of cycles and an initial surface,
as well as the system parameters, before we start the simulation.

```python
L = 100 #lattice units
T = 1.0 #J/kB
F = 1.0 #im units of r0 (rate constant)

N = 100000                                       #Number of cycles
surface_heights = np.random.randint(5, size=L) #start with a rough surface

all_surfaces, all_times = simulate_surface(surface_heights, L, N, T, F)
```

```python
#Visualize the surface evolution in time 
figsize(20, 10)
imshow(all_surfaces, extent=[0, L, 0, N], aspect="auto", origin="lower")
xlabel("site", fontsize=20)
ylabel("cycle", fontsize=20)
colorbar()

figure()
plot(all_surfaces[2000, :])

tight_layout()
show()
```

#Steady state average velocity vs. deposition rate F

```python
F_values = linspace(0.75, 1.25, 20)
T = 0.2 #J/kB

N = 30000                   #Number of cycles
thermalization = 10000      #Number of cycles before we start to measure

steady_state_speeds = []

for F in F_values:
    surface_heights = np.random.randint(5, size=L) #start with a rough surface
    all_surfaces, all_times = simulate_surface(surface_heights, L, N, T, F)
    
    average_heights = all_surfaces.mean(axis=1)
    
    delta_heigths = average_heights[thermalization+1:] - average_heights[thermalization]
    delta_times   = all_times[thermalization+1:] - all_times[thermalization]
    
    average_speeds = delta_heigths/delta_times
    
    steady_state_speed = average_speeds.mean()
    
    steady_state_speeds.append(steady_state_speed)
```

```python
from scipy.stats import linregress

#assume a linear relation and find the slope and shift of the linear fit
slope, shift, _, _, stderr = linregress(F_values, steady_state_speeds) 
linear_fit = [shift + slope*F for F in F_values]

#for a line y = ax+b, the x which yields y = 0 is x = -b/a
F0_estimate = -shift/slope

print "Estimated F0 = %.2f. Linear fit stderr = %g" % (F0_estimate, stderr)

plot(F_values, steady_state_speeds, "b*", label="KMC")
plot(F_values, linear_fit, "r--", label="linear fit") 

legend()
xlabel("F", fontsize=30)
ylabel("<v>", fontsize=30)

show()
```

#Surface roughness as a function of simulation time

The roughness is calculated as

$$w(h) = \sqrt{\langle (h - \langle h \rangle)^2 \rangle} = \sqrt{\frac{1}{L}\sum_{i=1}^L (h_i - \langle h \rangle)^2}$$

```python
def calculate_roughness(surface_heights):
    """
    Returns the roughness value of the given surface.
    """
    
    return std(surface_heights - surface_heights.mean())
```

A challange with measuring roughness as a function of time is fluctuations. In a single simulations, the fluctuations are very high.

We can cancel out some of this noise by taking the average over several time series. Note that since the time step is not fixed, 
summation will not correspond to the true average, but in the limit of infinite time series, this would be ok still.

We therefore define a function that generates such an average:

```python
def generate_average_roughness_vs_time(surface_heights, L, N, repeats, F, T):
    """
    Returns the roughness vs. time for [repeats] simulations 
    starting from the same initial configuration [surface_heights].
    """
    
    #Copy since a direct equal would link the memory of the two arrays
    surface_heights_orig = surface_heights.copy() 
    
    average_roughness = 0
    average_time = 0
    
    for repeat in range(repeats):
        
        surface_heights = surface_heights_orig.copy()
        all_surfaces, all_times = simulate_surface(surface_heights, L, N, T, F)
        
        all_roughnesses = [calculate_roughness(surface_heights) for surface_heights in all_surfaces]
        
        average_roughness += array(all_roughnesses) #convert to array to use the summation operator
        average_time += all_times
        
        #Fancy printout which updates itself instead of writing newlines
        sys.stdout.flush()
        print "\r%2d/%2d" % (repeat+1, repeats),
    print
    
    average_time /= repeats
    average_roughness /= repeats
    
    return average_roughness, average_time
```

```python
surface_heights = random.randint(5, size=L) #initial configuration

T = 0.2
F = 1.0
L = 50
N = 1500
repeats = 50

n_singles = 3

#first we do some single simulations to demonstrate the level of noise
for n in range(n_singles):
    average_roughness_single, average_time_single = generate_average_roughness_vs_time(surface_heights, L, N, 1, F, T)
    plot(average_time_single, average_roughness_single, label="single simulation %d" % n)
    
#then we do an average
average_roughness, average_time = generate_average_roughness_vs_time(surface_heights, L, N, repeats, F, T)
plot(average_time, average_roughness, label="averaged")

xlabel("time")
ylabel("roughness")

legend()
show()
```

#Data collapse of the roughness using the Family Viscek ansatz

This ansatz states that $w(t, L) \sim L^\alpha f(t/L^z)$. In order to fint values for $\alpha$, $z$, and $\beta = \alpha/z$, 
we compute the rougness as a function of time (as done previously), for various system sizes.

First, we calculate the roughness vs. time for various system sizes:

```python
L_values = [50, 100, 150, 200]

T = 0.2
F = 1.0
N = 5000
repeats = 100

times = []
average_roughnesses = []

for L in L_values:

    surface_heights = zeros(L)

    average_roughness, average_time = generate_average_roughness_vs_time(surface_heights, L, N, repeats, F, T)

    times.append(average_time)
    average_roughnesses.append(average_roughness)
```

```python
N = 5

z_values = linspace(0, 2, N)
alpha_values = linspace(0, 2, N)

figsize(20, 10)
f, axes_array = subplots(N, N)

for i, z in enumerate(z_values):
    for j, alpha in enumerate(alpha_values):
      
        for L, time, average_roughness in zip(L_values, times, average_roughnesses):    
            axes_array[i, j].plot(time/L**z, average_roughness/L**alpha, label="L = %3d" % L)
            axes_array[i, j].set_title("alpha = %g, z = %g" % (alpha, z))
xlabel("t")
ylabel("w(t, L)")
legend()
tight_layout()
show()
```

From the above coarse study, we find that both $\alpha=1/2$ and $z=1$ ($\beta = 1/2$), 
and $\alpha=1$ and $z = 2$ (also $\beta = 1/2$), are good values for data collapse.

```python
z0 = 1.
a0 = 0.5

z1 = 2.
a1 = 1.

figsize(20, 10)

figure()
for L, time, average_roughness in zip(L_values, times, average_roughnesses):    
    plot(time/L**z0, average_roughness/L**a0, label="L = %3d" % L)

xlim(0, 0.3)
title("alpha = %g, z = %g" % (a0, z0))
xlabel("t")
ylabel("w(t, L)")
legend()
show()


figure()
for L, time, average_roughness in zip(L_values, times, average_roughnesses):    
    plot(time/L**z1, average_roughness/L**a1, label="L = %3d" % L)

xlim(0, 0.003)
title("alpha = %g, z = %g" % (a1, z1))
xlabel("t")
ylabel("w(t, L)")
legend()
show()
```
